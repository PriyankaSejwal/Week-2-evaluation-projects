feature description:
    1. Province_State - The name of the state within the USA.
    2. Country_Region- The name of the country (i.e. USA).
    3. Last_Update- The recent date on which the data is updated.
    4. Lat- The latitude of the province_state.
    5. Long_- Longitude of the province state.
    6. Confirmed- Number of confirmed cases for covid 19.
    7. Deaths- Total death counts of the state for both confirmed and probable covid 19 cases.
    8. Recovered- Recovered counts of people from covid 19 for the state.
    9. Active- Confirmed cases that have not been resolved for the state. active=confirmed - recovered - total deaths.
    10. FIPS- Federal Information Processing Standards code to uniquely identify states in the USA.
    11. Incident_Rate- cases confirmed per 100,000 persons.
    12. People_Hopitalized- Total number of people hospitalized due to covid 19.
    13. Mortality_Rate- Number of recorded deaths*100/confirmed cases.
    14. UID- Unique identifier for each entry.
    15. ISO3- Official country code identifiers.
    16. Testing_Rate- Total number of people tested per 100,000.
    17. Hospitalization_Rate- total number of people hospitalized * 100/confirmed cases.

```python
#checking skewness of the data
us.skew()
```

```python
# checking for outliers
for i in us.columns:
    sns.boxplot(us[i])
    plt.show()
```

Outliers are present in all the columns. We willl treat the outliers with zscore.

```python
#Treating skewness using power transformer method method
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
for i in us.skew().index:
    if us.skew().loc[i]>0.55:
        us[i]=pt.fit_transform(us[i].values.reshape(-1,1))
    if us.skew().loc[i]<-0.55:
        us[i]=pt.fit_transform(us[i].values.reshape(-1,1))
    else:
        us[i]=us[i]
```

```python
us.skew()
```

```python
#Removing outliers using z score
from scipy.stats import zscore
z=np.abs(zscore(us))
```

```python
#Threshold=3
us_new=us[(z<3).all(axis=1)]
us_new
```

```python
print('before removing outliers shape was:', us.shape,'\t','after removing outliers shape is: ',us_new.shape)
```

```python
#splitting the data for model training
x=us_new.drop(['Deaths'],axis=1)
x
```

```python
#scaling using Min_Max Scaler
from sklearn.preprocessing import MinMaxScaler
mnm=MinMaxScaler()
```

```python
x1=mnm.fit_transform(x)
x=pd.DataFrame(x1,columns=x.columns)
x
```

```python
y=us_new['Deaths']
y
```

```python
x.shape
```

```python
y.shape
```

```python
#Modelling the data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
list=[LinearRegression(),SVR(),DecisionTreeRegressor(),KNeighborsRegressor()]
for j in list:
    print('for the algorithm: ',j)
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=59)
    j.fit(x_train,y_train)
    pred=j.predict(x_test)
    r2=r2_score(pred,y_test)
    print('Score is:',j.score(x_train,y_train))
    print('r2 score is: ',r2)
    print('Mean absolute error is: ',mean_absolute_error(pred,y_test))
    print('Mean squared error:',mean_squared_error(pred,y_test))
    print('Root mean squared error: ',np.sqrt(mean_squared_error(pred,y_test)))
    print("***********************************************************")
    print('\n')
```

```python
# Ridge, Lasso, ElasticNet
from sklearn.linear_model import Ridge,Lasso,ElasticNet
list=[Ridge(alpha=0.01),Lasso(alpha=0.01),ElasticNet(alpha=0.01)]
for j in list:
    print('for the algorithm: ',j)
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=59)
    j.fit(x_train,y_train)
    pred=j.predict(x_test)
    r2=r2_score(pred,y_test)
    print('Score is:',j.score(x_train,y_train))
    print('r2 score is: ',r2)
    print('Mean absolute error is: ',mean_absolute_error(pred,y_test))
    print('Mean squared error:',mean_squared_error(pred,y_test))
    print('Root mean squared error: ',np.sqrt(mean_squared_error(pred,y_test)))
    print("***********************************************************")
    print('\n')
```

```python
# Using ensemble techniques to boost up our score
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
list=[RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=59)
for m in list:
    m.fit(x_train,y_train)
    score=m.score(x_train,y_train)
    predict=m.predict(x_test)
    print('Score of :',m,"is:",score)
    print('r2 score is: ',r2)
    print('Mean absolute error is: ',mean_absolute_error(predict,y_test))
    print('Mean squared error:',mean_squared_error(predict,y_test))
    print('Root mean squared error: ',np.sqrt(mean_squared_error(predict,y_test)))
    print("***********************************************************")
    print('\n')
    
```

Observing all the algorithms we can conclude that ElasticNet is performing best with r2 score of 0.99 and mean absolute error of 0.039 and mean sqaured error of 0.002. Also ElasticNet regression avoids underftting and overfitting of the data.

```python
# Best parameters for ElasticNet
from sklearn.model_selection import GridSearchCV
en=ElasticNet()
parameters={ 'alpha':[0.001,0.01,0.1,1,10]}
cv=GridSearchCV(en,parameters)
cv.fit(x,y)
print(cv.best_params_)
```

```python
#working on elasticNet with alpha=0.001
enr=ElasticNet(alpha=0.001)
enr.fit(x_train,y_train)
score=enr.score(x_train,y_train)
predict=enr.predict(x_test)
print('Score is:',score)
print('r2 score is: ',r2)
print('Mean absolute error is: ',mean_absolute_error(predict,y_test))
print('Mean squared error:',mean_squared_error(predict,y_test))
print('Root mean squared error: ',np.sqrt(mean_squared_error(predict,y_test)))
```

```python
#applying cross validation to the best model to see if it is under fitting or over fitting
from sklearn.model_selection import cross_val_score
enr=ElasticNet(alpha=0.001)
cvs=cross_val_score(enr,x,y,cv=8)
print('The cross validation score is: ',cvs)
print('The mean and standard deviations are: ',cvs.mean(),cvs.std())
```

```python
#Saving the model
import joblib
joblib.dump(enr,'us_covid.pkl')
```
